```python
from pymystem3 import Mystem
import os
from string import punctuation
import nltk
```

```python
punctuation += '\n'
```

```python
with open ('lemfile.txt', 'w+', encoding='utf-8') as lemfile: 
    with open ('alice.txt', 'r+', encoding='utf-8') as alicetxt:
        m = Mystem()
        aliceread = alicetxt.read()
        alice1 = list(aliceread.split( ))
        alice2 = ''
        for i in alice1: 
            i = i.lower()
            i = str(i.strip(punctuation))
            alice2 += i 
            alice2 += ' '
        alicelemmatize = m.lemmatize(alice2)
        alicestr = ''
        for l in alicelemmatize:
            alicestr += l
        lemfile.write(alicestr)
print(alicestr)
alice3 = list(alicestr.split( ))
```
