## Задание 1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
from pymystem3 import Mystem
import os
from string import punctuation
```

```python
punctuation += '\n' #добавляю в пунктуацию, которую буду отрезать, обозначение переноса строки
def funct(a, b): #
    with open(a, 'w+', encoding='utf-8') as lemfile:
        with open(b, 'r+', encoding='utf-8') as la:
            m = Mystem()
            laread = la.read()
            lemmas_la = m.lemmatize(laread) #лемматизирую строку из прочитанного файла b
            strlem = '' #создаю пустую строковую переменную, в которую буду добавлять строковые элементы с обрезанной пунктуацией
            for l in lemmas_la: 
                l = str(l.strip(punctuation)) #отрезаю пунктуацию 
                strlem += l
            lemfile.write(strlem) #записываю содержимое строковой переменной в файл a
```

```python
lemmas = funct('lemmas_file.txt', 'literary_anecdotes.txt') #прописаны пути к функции
```

## Задание 2

Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. 

```python
with open('lemmas_file.txt', 'r+', encoding='utf-8') as lf:
    lfile = lf.read()
    lemlist = list(lfile.split(' ')) #делаю список из строгового содержимого файла; делю элементы по пробелам
    newlist = []
    with open('rus_stopwords.txt', 'r+', encoding='utf-8') as stpwrds:
        stopwords = stpwrds.read()
        stoplist = list(stopwords.split('\n')) #делаю список из строгового содержимого файла; делю элементы по знакам переноса строки
        for lem in lemlist:
            lem = lem.lower() #привожу все леммы к нижнему регистру
            if lem not in stoplist:
                if lem.isalpha(): #исключаю все леммы, которые не являются словами, чтобы избавиться от "-"
                    newlist.append(lem) #добавляю леммы в список 
        print(newlist)
```

Выведите 20 самых частотных по этому параметру слов.

```python
ipmdict = {}
for a in newlist:
    ipm = newlist.count(a) / len(newlist) * 1000000 #считаю ipm
    ipmdict.update({a:ipm}) #добавляю слово и соответствующее значение ipm в словарь; слово - ключ. 
```

```python
sorted_ipmdict = sorted(ipmdict.items(), key=lambda x: x[1], reverse=True) #сортирую содержимое словаря
```

```python
print(sorted_ipmdict[:20]) #вывожу первые 20 элементов: это будет 20 слов с наибольшим ipm
```
